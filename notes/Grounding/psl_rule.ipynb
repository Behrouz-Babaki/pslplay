Suppose that we want to add the following ground rule to the mathematical program:
    
\begin{equation}
    B_1 \land B_2 \land \ldots \land B_n \to H
\end{equation}

Assume that $b_1 \ldots b_n$ are the variables (or constants) corresponding to atoms $B_1 \ldots B_n$. Similarly $h$ corresponds to ground atom $H$. Now if we represent the distance to satisfaction for this rule by a variable $y$ in the linear program, the following constraint is enough for establishing the connection between these variables:

\begin{equation}
    -\sum_{i=1}^n (1-b_i) -h -y + 1 \leq 0
\end{equation}

The reason lies in the associative property of $\land$ and $\lor$ operators in PSL. Recall that $a \tilde{\lor} b = \min \{1, a+b\}$. So we have:

\begin{equation}
    a \tilde{\lor} b \tilde{\lor} c = \min \{ 1, \min \{1, a+b\}+c \} = \min \{1, c+1, a+b+c \}
\end{equation}

However, $c \in [0,1] \to c+1 \geq 1$. So  We have $a\,\tilde{\lor}\,b\,\tilde{\lor}\,c\,\tilde{\lor} = \min \{1, a+b+c \}$, and in general:

\begin{equation}
       \tilde{\bigvee}_{i=1}^n a_i = \min \{ 1, \sum_{i=1}^n a_i \}
\end{equation}

We can show that a similar equation holds for the $\land$ operator:

\begin{equation}
       \tilde{\bigwedge}_{i=1}^n a_i = \max \{ 0, \sum_{i=1}^n a_i - (n-1) \}
\end{equation}


Now we can compute the interpretation of ground rule $B_1 \land B_2 \land \ldots \land B_n \to H$ as follows:

\begin{equation}
    I(B_1 \land B_2 \land \ldots \land B_n \to H) = \\
    I(\neg B_1 \lor \neg B_2 \lor \ldots \neg \lor B_n \lor H) = \\
    \min \{ \sum_{i=1}^n (1-I(B_i)) + I(H), 1 \}
\end{equation}

And the distance to satisfaction (i.e. $1 - I(G_R)$) will be:

\begin{equation}
    1 - \min \{ \sum_{i=1}^n (1-I(B_i)) + I(H), 1 \} = \\
    \max \{1 - \sum_{i=1}^n (1-I(B_i)) - I(H), 0 \}
\end{equation}

which will be encoded by the constraint:

\begin{equation}
    y \geq 1 -\sum_{i=1}^n (1-b_i) -h
\end{equation}

The function *psl_rule* creates variable $y$, and adds the above-mentioned linear constraint to the linear program:

```python
def psl_rule(problem, body, head, body_neg_stat, head_neg_stat):
    y = problem.add_var()
    var_ids = [y]
    coefs = [-1]
    const_part = 1
    
    isconst, val = head
    if isconst:
        if head_neg_stat:
            const_part += (-1+val)
        else:
            const_part += -val
    else:
        var_ids.append(val)
        if head_neg_stat:
            const_part -= 1
            coefs.append(1)
        else:
            coefs.append(-1)
            
    for i in range(len(body)):
        isconst, val = body[i]
        if isconst:
            if body_neg_stat:
                const_part -= val
            else:
                const_part -= (1-val)
        else:
            var_ids.append(val)
            if body_neg_stat:
                coefs.append(-1)
            else:
                coefs.append(1)
                const_part -= 1
    problem.add_linear_constraint(var_ids, coefs, const_part)  
    return (False,y)
```
